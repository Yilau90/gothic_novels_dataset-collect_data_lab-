# Find out authors who are male and born in United Kingdom

```python
novel_df[novel_df['country_of_birth'].str.contains('United Kingdom') & novel_df['gender'].str.contains('Male')]
```

```python
uk_male_novels_df=novel_df[novel_df['country_of_birth'].str.contains('United Kingdom') & novel_df['gender'].str.contains('Male')]
```

# Find out uk male authors' novels, group by novel names and sorted by publication year

```python
uk_male_novels_df.groupby('novel_name')[['publication_year']]\
.sum().sort_values(by='publication_year', ascending=False)
```

# Find out which country most female authors come from and visualize it in a pie chart

```python
women_filter = novel_df['gender'] == 'Female'
```

```python
novel_women = novel_df[women_filter]
```

```python
novel_women['country_of_birth'].value_counts()
```

```python
novel_women['country_of_birth'].value_counts().plot(kind='pie', figsize=(10,10), title='Gothic novels:\nMost Frequent "Coutry of birth" Among Female authors')
```
